```python
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine

import getpass  
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
query1 = 'SELECT COUNT(DISTINCT(last_name)) AS different_actor_names FROM actor'


dif_act = pd.read_sql_query(query1, engine)
dif_act
```

```python
query2 = '''SELECT * ,DATE_FORMAT(rental_date, '%%M') AS 'month' ,DATE_FORMAT(rental_date, '%%W') AS 'weekday' ,CASE WHEN DATE_FORMAT(rental_date, '%%W') > 5 THEN 'weekend' ELSE 'workday' END AS day_type FROM rental ORDER BY rental_date;'''
wkday = pd.read_sql_query(query2, engine)
wkday.head()
```

```python
query3= 'SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"'
arm = pd.read_sql_query(query3, engine)
arm.head()
```

```python
query4= 'SELECT title, length FROM film ORDER BY length DESC LIMIT 10'
tp_10 = pd.read_sql_query(query4, engine)
tp_10
```

```python
query5= 'SELECT * FROM film WHERE special_features LIKE "%%Behind the Scenes%%"'
bh_sc = pd.read_sql_query(query5, engine)
bh_sc.head()
```

```python
query6= 'SELECT rating, ROUND(AVG(length),2) as "AVG length" FROM film GROUP BY rating HAVING AVG(length)>120'
hrs_2 = pd.read_sql_query(query6, engine)
hrs_2.head()
```

```python
query7= 'SELECT title, length,RANK () OVER (ORDER BY length DESC) AS "rank" FROM film WHERE length IS NOT NULL'
length = pd.read_sql_query(query7, engine)
length.head()
```
